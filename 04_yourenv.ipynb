### The Anatomy of a Gymnasium Environment



To make your environment compatible with standard RL algorithms, it **must** inherit from `gym.Env` and implement these core methods:

* **`__init__(self)`**: The constructor. You *must* define `self.observation_space` and `self.action_space` here using `gym.spaces`.
* **`reset(self, seed=None)`**: Called to start a new episode. It must reset the internal state and return the initial observation and an empty info dictionary: `return obs, {}`
* **`step(self, action)`**: The core physics engine. It takes an action, updates the internal state, calculates the reward, and checks if the episode is over. It must return a tuple of exactly 5 elements: `(obs, reward, terminated, truncated, info)`.
