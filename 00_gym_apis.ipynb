### 1. The Core Gymnasium API

To build a Reinforcement Learning (RL) loop, we first need to understand how the agent interacts with the environment. Gymnasium provides a simple, unified interface to facilitate this:

* **`gym.make(id)`**: Initializes and creates the environment.
* **`env.reset()`**: Starts a new episode and returns the initial observation (state).
* **`env.step(action)`**: Executes an action in the environment. It returns a tuple containing:
    * `observation` (array): The new state of the environment.
    * `reward` (float): The numerical reward received from taking the action.
    * `terminated` (bool): Whether a terminal state is reached (e.g., the pole falls over).
    * `truncated` (bool): Whether the episode was artificially ended (e.g., hitting a time limit).
    * `info` (dict): Additional diagnostic information.
* **`env.close()`**: Closes the environment and safely frees up resources.

---

### 2. The Reinforcement Learning Loop

The core of any RL implementation is the continuous interaction loop between the **Agent** and the **Environment**.



The process follows a straightforward cycle:
1. The environment provides the agent with an initial **observation**.
2. The agent decides on an **action** to take.
3. The environment steps forward, returning the **next observation** and a **reward**.
4. This repeats until the episode is `terminated` or `truncated`.
